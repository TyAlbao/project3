```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
raw = pd.read_csv('mouse_data.csv')
raw['mins'] = raw.index
```

```python
df = pd.melt(raw, id_vars = ['mins'], value_vars= raw.columns).sort_values(['mins', 'variable'])
df['id'] = df['variable'].apply(lambda x: x.split('_')[0])
df['variable'] = df['variable'].apply(lambda x: x.split('_')[-1])
df['sex'] = df['id'].apply(lambda x: x[0])
df = df[['mins', 'sex', 'id', 'variable', 'value']]
df.sort_values(['id', 'variable'])
df = df.reset_index(drop = True)
df
```

```python
avg_temp_f = df[(df['sex'] == 'f') & (df['variable'] == 'temp')].groupby('mins')[['value']].mean()
avg_temp_m = df[(df['sex'] == 'm') & (df['variable'] == 'temp')].groupby('mins')[['value']].mean()

avg_act_f = df[(df['sex'] == 'f') & (df['variable'] == 'act')].groupby('mins')[['value']].mean()
avg_act_m = df[(df['sex'] == 'm') & (df['variable'] == 'act')].groupby('mins')[['value']].mean()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(y = avg_temp_f['value'], name = 'female', line_color = 'maroon'))
fig.add_trace(go.Scatter(y = avg_temp_m['value'], name = 'male', line_color = 'skyblue'))

total_mins = df['mins'].nunique()
light_times = range(0, total_mins, 720)
for i in range(len(light_times)):
    if i % 2 == 0:
        color = 'black'
    else:
        color = 'grey'
    fig.add_vline(x = light_times[i], line_color = color)

fig.show()
```
